```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen en modo de escala de grises
image_gray = cv2.imread('bariloche2/B1.tif', cv2.IMREAD_GRAYSCALE)

# Verificar si la imagen se ha cargado correctamente
if image_gray is None:
    print("Error al cargar la imagen.")
else:
    # Función para ecualizar histograma en una imagen en escala de grises
    def equalize_histogram(image):
        return cv2.equalizeHist(image)

    # Aplicar ecualización en la imagen en escala de grises
    equalized_gray = equalize_histogram(image_gray)

    # Mostrar la imagen original y la ecualizada
    plt.figure(figsize=(12, 6))

    # Imagen original en escala de grises
    plt.subplot(1, 2, 1)
    plt.imshow(image_gray, cmap='gray')
    plt.title("Imagen Original en Escala de Grises")
    plt.axis('off')

    # Imagen ecualizada
    plt.subplot(1, 2, 2)
    plt.imshow(equalized_gray, cmap='gray')
    plt.title("Imagen Ecualizada en Escala de Grises")
    plt.axis('off')

    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen en modo de escala de grises
image_gray = cv2.imread('bariloche2/B1.tif', cv2.IMREAD_GRAYSCALE)

# Verificar si la imagen se ha cargado correctamente
if image_gray is None:
    print("Error al cargar la imagen.")
else:
    # Función para ecualizar histograma en una imagen en escala de grises
    def equalize_histogram(image):
        return cv2.equalizeHist(image)

    # Aplicar ecualización una vez
    equalized_once = equalize_histogram(image_gray)

    # Aplicar ecualización una segunda vez sobre la imagen ya ecualizada
    equalized_twice = equalize_histogram(equalized_once)

    # Mostrar la imagen original, ecualizada una vez y ecualizada dos veces
    plt.figure(figsize=(18, 6))

    # Imagen original
    plt.subplot(1, 3, 1)
    plt.imshow(image_gray, cmap='gray')
    plt.title("Imagen Original en Escala de Grises")
    plt.axis('off')

    # Imagen ecualizada una vez
    plt.subplot(1, 3, 2)
    plt.imshow(equalized_once, cmap='gray')
    plt.title("Imagen Ecualizada una vez")
    plt.axis('off')

    # Imagen ecualizada dos veces
    plt.subplot(1, 3, 3)
    plt.imshow(equalized_twice, cmap='gray')
    plt.title("Imagen Ecualizada dos veces")
    plt.axis('off')

    plt.show()
```

Al intentar ecualizar por segunda vez no hay un cambio significativo, el algoritmo no puede expandir ni ajustar más el rango de intensidades.
