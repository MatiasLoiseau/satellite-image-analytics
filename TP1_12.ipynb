```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Función para aplicar ruido de Sal y Pimienta
def agregar_ruido_sal_pimienta(imagen, densidad):
    # Crear una copia de la imagen
    imagen_con_ruido = np.copy(imagen)

    # Calcular la cantidad de píxeles a afectar
    total_pixels = imagen.size
    num_salt = int(total_pixels * densidad / 2)
    num_pepper = int(total_pixels * densidad / 2)

    # Agregar ruido de 'Sal' (blancos)
    coords_salt = (np.random.randint(0, imagen.shape[0], num_salt),
                   np.random.randint(0, imagen.shape[1], num_salt))
    imagen_con_ruido[coords_salt] = 255

    # Agregar ruido de 'Pimienta' (negros)
    coords_pepper = (np.random.randint(0, imagen.shape[0], num_pepper),
                     np.random.randint(0, imagen.shape[1], num_pepper))
    imagen_con_ruido[coords_pepper] = 0

    return imagen_con_ruido

# Cargar imagen en escala de grises
imagen = cv2.imread("bariloche2/B1.tif", cv2.IMREAD_GRAYSCALE)

# Densidad de ruido
densidad = 0.05  # Ajusta este valor entre 0 y 1 para cambiar la densidad del ruido

# Aplicar ruido de Sal y Pimienta
imagen_ruidosa = agregar_ruido_sal_pimienta(imagen, densidad)

# Visualización de la imagen original y la imagen con ruido
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_ruidosa, cmap='gray')
plt.title(f'Imagen con Ruido Sal y Pimienta (Densidad={densidad})')
plt.axis('off')

plt.show()
```
