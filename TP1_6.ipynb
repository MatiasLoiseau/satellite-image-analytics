```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def negativo_rgb(image_path):
    # Cargar la imagen en formato BGR
    imagen_bgr = cv2.imread(image_path)

    # Verificar si la imagen se carg√≥ correctamente
    if imagen_bgr is None:
        print(f"No se pudo cargar la imagen en la ruta: {image_path}")
        return None

    # Convertir la imagen de BGR a RGB
    imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)

    # Calcular el negativo de la imagen en RGB
    imagen_negativa = 255 - imagen_rgb

    return imagen_rgb, imagen_negativa

# Ruta de la imagen de entrada
ruta_imagen = 'bariloche2/B1.tif'
```

```python
# Obtener la imagen original y su negativo
imagen_original, imagen_negativa = negativo_rgb(ruta_imagen)

# Mostrar la imagen original y la negativa con matplotlib
if imagen_original is not None and imagen_negativa is not None:
    plt.figure(figsize=(10, 5))

    # Mostrar imagen original
    plt.subplot(1, 2, 1)
    plt.imshow(imagen_original)
    plt.title("Imagen Original")
    plt.axis('off')

    # Mostrar imagen negativa
    plt.subplot(1, 2, 2)
    plt.imshow(imagen_negativa)
    plt.title("Imagen Negativa")
    plt.axis('off')

    plt.show()
```
